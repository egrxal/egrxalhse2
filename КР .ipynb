### Задание 9_Факторный анализ (1 балл)

```python
heart = pd.read_csv('heart.csv')
heart
```

Ищем предпосылки для проведения факторного анализа - как минимум должны быть корреляции между переменными

```python
correlation = heart.corr()
```

```python
#Построим матрицу корреляция и разукрасим, чтобы не было скучно

fig , ax = plt.subplots(figsize = (20,16))

sns.heatmap(correlation, 
            linewidths=0.5, annot=True,cmap='viridis',
            linecolor="white", annot_kws = {'size':12})
```

Очевидно, корреляции присутствуют, но не во всех случаях нам подходят переменные для проведения факторного анализа - как минимум нужно убрать возраст и аналогичные переменные - они для факторного анализа не подойдут. Категориальные переменные тоже в топку.

```python
df = heart.drop(['age', 'sex', 'cp','restecg','thal','target', 'exang'], axis=1)
df
```

```python
correlation = df.corr()
```

```python
#Построим матрицу корреляция и разукрасим, чтобы не было скучно

fig , ax = plt.subplots(figsize = (20,16))

sns.heatmap(correlation, 
            linewidths=0.5, annot=True,cmap='viridis',
            linecolor="white", annot_kws = {'size':12})
```

Как видно, корреляции у нас небольшие, но давайте посмотрим на показатели

```python
from factor_analyzer import FactorAnalyzer
```

```python
df.head()
```

```python
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
chi_square_value,p_value=calculate_bartlett_sphericity(df)
chi_square_value, p_value
```

Значимость критерия Бартлетта меньше 0,05 указывает, что данные приемлемы для проведения факторного анализа

```python
from factor_analyzer.factor_analyzer import calculate_kmo
kmo_all,kmo_model=calculate_kmo(df)
```

```python
kmo_all
```

```python
fa = FactorAnalyzer(rotation = 'varimax')
fa.fit(df)

ev, v = fa.get_eigenvalues()
ev
```

Как видно, нам подойдут 2 или 3 переменные (но скорее 2)

```python
plt.scatter(range(1, df.shape[1] + 1), ev)
plt.plot(range(1, df.shape[1] + 1), ev)

plt.title('График собственных значений')
plt.xlabel('Признаки')
plt.ylabel('Собственные значения')
```

```python
fa = FactorAnalyzer (n_factors = 2, rotation = 'varimax')
fa.fit(df)
```

```python
loadings = fa.loadings_
```

```python
loadmatrix = pd.DataFrame.from_records(loadings, index = df.columns, columns= ['Fac1', 'Fac2'])
loadmatrix
```

```python
import numpy as np
```

```python
loadmatrix1 = loadmatrix.where(loadmatrix > 0.1, np.nan)
loadmatrix1
```

```python
loadmatrix2 = loadmatrix1.sort_values((['Fac1','Fac2']), ascending=False)
```

```python
loadmatrix2
```

```python
plt.figure(figsize = (14,14))
ax = sns.heatmap(loadmatrix2, annot = True, cmap = 'BuPu')

ax.yaxis.set_tick_params(labelsize = 15)
plt.title('Факторный анализ', fontsize = 'xx-large')

plt.ylabel('Диапазон', fontsize = 'xx-large')
```

В результате факторного анализа выделены два фактора - в первый фактор выделены наклон пикового значения в электрокардиограмм и максимальный пульс, что логично, так как оба связаны с интенсивными управжнениями. Второй фаткор преимущественно объединяет различные показатели - сахар в крови и др, что также логично.
